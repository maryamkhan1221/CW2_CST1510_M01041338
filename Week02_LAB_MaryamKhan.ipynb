4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese

Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
weight_pounds= float(input("Enter your weight in pounds: ")) #ask user to input their weight
height_inches= float(input("Enter your height in inches: ")) #ask uset to input their height

weight_kg= weight_pounds* 0.45359237 #convert pounds to kg
height_m= height_inches* 0.0254 #convert inches to meters

bmi= weight_kg / (height_m** 2) #calculate BMI

print("Your BMI is:", round(bmi, 2)) #round to 2 decimal places

if bmi <18.5:
    print("You are underweight.") #BMI< 18.5
elif bmi <= 25.0:
    print("You are normal weight.") #BMI between 18.5 and 25
elif bmi <= 30.0:
    print("You are overweight.") #BMI between 25 and 30
else:
    print("You are obese.") #BMI over 30
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
marks= [89, 65, 34, 54, 89, 76, 42] #list of student marks

total= 0
for mark in marks:
    total+= mark #add each mark to total

average=total/ len(marks) #calculate average

highest= max(marks) #find highest mark
lowest= min(marks) #find lowest mark

passed= 0
failed= 0

for mark in marks:
    if mark>= 50:
        passed+= 1 #count students who passed
    else:
        failed+= 1 #count students who failed

print("Average mark:", round(average, 2)) #show average
print("Highest mark:", highest) #show highest
print("Lowest mark:", lowest) #show lowest
print("Number of students who passed:", passed) #show passed count
print("Number of students who failed:", failed) #show failed count
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
secret = 7  #secret number to guess
number = None  #initialize user guess
attempts = 0  #initialize attempts counter

while number != secret:  #repeat until correct guess
    number = int(input("Enter your guess: "))  #get guess from user
    attempts += 1  #increase attempts

    if number < secret:  #too low
        print("Too low!")  
    elif number > secret:  #too high
        print("Too high!")  

print("Correct! You guessed it in", attempts, "tries.")  #correct guess
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
>


---

````python
numbers= [] #empty list to store numbers

while True: #loop until user types done
    user_input= input("Enter a positive number or type 'done' to stop: ").lower() #get input and make lowercase

    if user_input== "done": #if user wants to stop
        break

    if user_input.isdigit(): #check if input is a number
        number= int(user_input) #convert to integer
        if number> 0: #only store positive numbers
            numbers.append(number)
        else:
            print("Please enter a positive number.")
    else:
        print("Invalid input. Enter a number or 'done'.")

if numbers: #if list not empty
    print("Numbers entered:", numbers) #show all numbers

    even_found= False #flag to check for even number
    for num in numbers:
        if num % 2== 0: #check for even
            print("First even number:", num)
            even_found= True
            break
    if not even_found:
        print("No even numbers found")
else:
    print("No numbers were entered")
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
import random  #import random module

lottery = random.randint(10, 99)  #generate random 2-digit lottery number
guess = int(input("Enter your two-digit number: "))  #user input

lottery_tens = lottery // 10  #get tens digit of lottery
lottery_ones = lottery % 10  #get ones digit of lottery
guess_tens = guess // 10  #get tens digit of guess
guess_ones = guess % 10  #get ones digit of guess

print("Lottery number:", lottery)  #show lottery number

if guess == lottery:  #exact match
    print("Correct, exact match you win £10,000!")  
elif (guess_tens == lottery_ones and guess_ones == lottery_tens):  #digits reversed
    print("All digits match you win £3,000!")  
elif (guess_tens == lottery_tens or guess_tens == lottery_ones or
      guess_ones == lottery_tens or guess_ones == lottery_ones):  #one digit match
    print("One digit matches, you win £1,000!")  
else:  #no match
    print("Sorry, no match.")  
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
num_students= int(input("Enter the number of students: ")) #get number of students

highest= -1 #store highest score
second_highest= -1 #store second highest score

for i in range(num_students): #loop through all students
    score= float(input("Enter student score: "))  #get each score

    if score> highest: #if new score is higher than highest
        second_highest= highest #move old highest to second highest
        highest= score #update highest with new score
    elif score> second_highest and score != highest: #if score is between highest and second
        second_highest= score  #update second highest

print("Highest score:", highest) #print highest score
print("Second highest score:", second_highest) #print second highest score
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
# Ask the user to enter the actual password
actual= input("Enter actual password: ")

# Ask the user to enter their guessed password
guess= input("Enter guessed password: ")

# Find the maximum length between the two passwords
length= max(len(actual), len(guess))

# Start a counter for matching characters
matches= 0

# Compare each character one by one
for i in range(length):
    a= actual[i] if i < len(actual) else None  # Get character from actual password if it exists
    g= guess[i] if i < len(guess) else None    # Get character from guessed password if it exists
    if a is not None and g is not None and a == g:  # Check if both characters match exactly (same case and position)
        matches += 1  # Increase the count for matching characters

# Calculate accuracy as a percentage
accuracy = (matches / length) * 100

# If accuracy is 100%, print "Password cracked!"
if accuracy== 100:
    print("Password cracked!")
else:
    # Otherwise, show how accurate the guess was
    print(f"Accuracy: {round(accuracy)}%")
````
