3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
def password_to_ascii(password):
    result = ""
    for char in password:
        result += str(ord(char))
    return result

password = input("Enter your password: ")
ascii_password = password_to_ascii(password)
print("Password in ASCII:", ascii_password)
````

4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def check_password(password):
    has_upper = any(character.isupper() for character in password)
    has_lower = any(character.islower() for character in password)
    has_digit = any(character.isdigit() for character in password)
    has_special = any(character in "@#$%^&*!" for character in password)
    has_space = " " in password

    if has_space:
        print("Password should not contain spaces.")
    else:
        print("Uppercase:", has_upper)
        print("Lowercase:", has_lower)
        print("Digit:", has_digit)
        print("Special character:", has_special)

password = input("Enter your password: ")
check_password(password)
````

5. Create a function validating a student number. The function takes one parameter for a student number and returns True if the student number starts with “M00”, and the student number has 9 characters

````python
def validate_student_number(student_number):
    return student_number.startswith("M00") and len(student_number) == 9

student_number = input("Enter your student number: ")
if validate_student_number(student_number):
    print("Valid student number.")
else:
    print("Invalid student number.")
````

7. Create a function generating an email for an organization: Ask the user for his or her initials. Make sure that there is no whitespace added to the initials Before adding the initials make sure that there are lower case letters Concatenate the initial with a randomly generated three-digit number and the ‘@live.mdx.ac.uk’. Return the new email.

````python
import random

def generate_email():
    initials = input("Enter your initials: ").strip().lower()
    random_number = random.randint(100, 999)
    email = initials + str(random_number) + "@live.mdx.ac.uk"
    return email

print(generate_email())
````
