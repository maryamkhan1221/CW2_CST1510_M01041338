### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
seasons = 5  # there are 5 seasons in total
episodes_per_season = 10  # each season has 10 episodes
season1_time = (45 * 60) + 35  # season 1 episode = 45 mins and 35 secs, converted to seconds
season2_5_time = (35 * 60) + 15  # seasons 2-5 episodes = 35 mins and 15 secs, converted to seconds
total_seconds = season1_time * episodes_per_season + season2_5_time * episodes_per_season * 4  # total time for all episodes
hours = total_seconds // 3600  # convert seconds to hours
minutes = (total_seconds % 3600) // 60  # leftover minutes after hours
seconds = total_seconds % 60  # leftover seconds after minutes
print("Total time:", hours, "h", minutes, "m", seconds, "s")  # show the total watch time
````

### Typical 2. Savings Projection (3 Years)

Compute final savings after 3 years given:
- Save £10 every week for 3 years
- Year 1 bonus: +£20 each week
- Every 2 weeks: −£0.50 (sister)
Assume 52 weeks per year; use arithmetic and augmented assignment; no conditionals.

**Example Output:**
```
Final savings after 3 years: £4689.50
```


````python
weeks=52*3  #3 years, each year has 52 weeks
weekly_save=10  #saving £10 every week
bonus=20  #extra £20 every week in year 1
sister_cut=0.5/2  #sister takes 50p every 2 weeks so 25p per week.
total_savings=(weekly_save+bonus-sister_cut)*weeks  #total after 3 years
print("Final savings after 3 years: £",total_savings)  #show final money
````

### Typical 4. Data Science: Dataset Metadata
